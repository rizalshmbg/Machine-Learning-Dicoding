# Pengenalan Natural Language Processing (NLP)

Bayangkan Anda terbangun di pagi hari. Dengan mata terpejam, mengaktifkan perangkat cerdas di sisi tempat tidur kemudian bertanya pada asisten digital:

Anda : Jam berapa sekarang?
Asisten digital : Sekarang pukul enam pagi

Tidak seperti biasanya, pagi ini Anda merasa sangat kedinginan. Kemudian, sambil menarik selimut, Anda bertanya lagi:

Anda : Berapa suhu saat ini?
Asisten digital : Sekarang, suhu di Bandung 14

Dengan suhu sedingin ini, rasanya enggan untuk beranjak dari tempat tidur. Melanjutkan tidur sambil meringkuk di dalam selimut yang hangat sepertinya ide bagus. Namun, Anda kemudian ingat ada janji temu klien hari ini tapi entah jam berapa. Anda lantas bertanya lagi:

Anda: Apa saja jadwalku hari ini?

Asisten digital menjawab dengan membacakan seluruh jadwal pada kalender digital Anda hari itu, termasuk jadwal meeting dengan klien pada pukul 7 pagi. Artinya, Anda tidak bisa lanjut rebahan dan harus segera bangkit dari tempat tidur untuk bersiap.

Anda mungkin familiar dan pernah menggunakan asisten digital. Bahkan, sebagian dari Anda mungkin telah lama menggunakannya dalam kehidupan sehari-hari. Tentu saja, bahasa yang Anda gunakan saat berkomunikasi dengan asisten digital bukanlah bahasa pemrograman formal seperti Python atau C, melainkan bahasa alami. Bahasa alami yang Anda gunakan sehari-hari untuk berkomunikasi dengan sesama manusia.

Permasalahannya, komputer hanya dapat memproses data dalam biner. Lalu, bagaimana kita membuat mesin memahami bahasa alami tersebut? Di sinilah bidang **Natural Language Processing** (Pemrosesan Bahasa Alami) berperan. NLP merupakan subbidang dari Artificial Intelligence (AI) untuk menganalisis, memodelkan, dan memahami bahasa manusia [30](https://livebook.manning.com/book/real-world-natural-language-processing/chapter-1/v-8/). NLP termasuk subbidang AI karena pemrosesan bahasa dianggap sebagai bagian dari kecerdasan manusia. Penggunaan bahasa merupakan keterampilan paling menonjol yang membedakan manusia dengan hewan lainnya.

NLP mencakup berbagai algoritma, metode, dan permasalahan yang menggunakan teks sebagai input. NLP menghasilkan beberapa informasi seperti label, representasi semantik, dan sebagainya, sebagai output. Teknik NLP digunakan di setiap aplikasi cerdas yang melibatkan bahasa alami. Ia merupakan komponen penting dalam berbagai aplikasi perangkat lunak yang kita gunakan dalam kehidupan sehari-hari [31](http://www.practicalnlp.ai/).

Selanjutnya, kita akan mempelajari beberapa aplikasi NLP dalam dunia nyata serta beberapa contoh kasus yang menggunakan metode NLP dalam penyelesaiannya.

# Aplikasi Natural Language Processing di Dunia Nyata

Di era digital yang serba modern, sebagian besar komunikasi dilakukan secara online. Tentu saja komunikasi online ini menggunakan bahasa alami, entah itu dalam bentuk teks, maupun suara. Selama komunikasi online ini ada, maka kebutuhan untuk pengembangan NLP juga akan tetap ada. 

Besarnya peluang pengembangan NLP ini berdampak pada semakin banyaknya perusahaan menerapkan NLP dalam produk maupun bisnisnya. Hal ini berpengaruh pada permintaan akan pakar atau tenaga ahli di bidang NLP. Tentu ini peluang bagus bagi Anda yang tertarik untuk mendalami NLP. Semangat!

Nah, berbicara tentang penerapan NLP, kita bisa menemukan penerapan NLP pada banyak bidang. Sebagai contoh, bidang e-commerce, pendidikan, keuangan, hiburan, perawatan kesehatan, layanan pelanggan, dan marketing. Pada bidang marketing, salah satu aplikasi NLP yang telah dibahas di materi sebelumnya adalah analisis sentimen untuk social media monitoring. Selain yang telah disebutkan tadi, NLP memiliki peran signifikan pada berbagai produk dan layanan. Berikut adalah beberapa aplikasi NLP yang paling populer.

*  Mesin penerjemah : 
Layanan penerjemah merupakan aplikasi langsung dari NLP. Ia adalah salah satu bidang yang paling banyak dipelajari dan termasuk aplikasi NLP yang paling awal dikembangkan. Sistem ini membantu kita menerjemahkan teks dari satu bahasa ke bahasa lain dengan mudah. Pada mesin penerjemah, bahasa untuk masukan disebut bahasa sumber, sedangkan bahasa untuk keluaran disebut bahasa target. Mesin penerjemah mencakup berbagai teknik NLP. Untuk memahami bahasa masukan, mesin penerjemah menggunakan teknik pemahaman bahasa. Sedangkan untuk menghasilkan keluaran atau hasil terjemahan, ia menggunakan teknik language generation (teknik untuk menghasilkan bahasa).
Jaringan internet memungkinkan perusahaan memiliki klien dan memasarkan produk ke seluruh penjuru dunia. Hal ini berimbas pada kebutuhan untuk menerjemahkan banyak dokumen dalam berbagai bahasa. Mesin penerjemah mengambil peran penting dalam proses ini. Layanan streaming seperti Amazon, Netflix, dan YouTube juga menggunakan mesin penerjemah untuk menghasilkan terjemahan dalam berbagai bahasa. Selain itu, mesin penerjemah yang tersedia secara daring seperti Google Terjemahan membantu wisatawan di seluruh dunia berkomunikasi dalam bahasa lokal.

Berikut adalah contoh bagaimana mesin penerjemah dapat menerjemahkan tulisan dalam gambar secara real-time [32](https://blog.google/products/translate/google-translates-instant-camera-translation-gets-upgrade/)

*  Digital Assistant
Contoh selanjutnya, asisten digital. Asisten berbasis suara seperti Apple Siri, Google Assistant, Microsoft Cortana, Amazon Alexa, dan Samsung Bixby mengandalkan berbagai teknik NLP untuk berinteraksi dengan pengguna. Contoh interaksinya misal, memahami perintah pengguna kemudian merespon dengan tepat.
Asisten digital merupakan tipe sistem dialog yang berorientasi pada tugas tertentu (task-oriented dialog system). Sistem ini ada untuk mencapai berbagai tujuan atau goals, misalnya memperoleh informasi, mengatur jadwal di kalender, mencari restoran terdekat, memesan tiket perjalanan, dsb. Ia dibangun sebagai NLP pipeline yang terdiri dari beberapa komponen, antara lain: pengenalan suara (speech recognition), pemahaman bahasa (language understanding), manajemen dialog, produksi respons, dan sintesis suara. Semua komponen ini biasanya dilatih secara terpisah dari hulu ke hilir (end-to-end) [30](https://livebook.manning.com/book/real-world-natural-language-processing/chapter-1/v-8/).

*  Mesin Pencari
Siapa yang tidak mengenal mesin pencari? Tempat kita mencari dan bertanya mengenai berbagai hal di internet. Mesin pencari modern seperti Google dan Bing seolah telah menjadi bagian tak terpisahkan dalam kehidupan daring umat manusia. Nah, dibalik kecanggihan mesin pencari, ada peran penting NLP dalam berbagai tahap pengembangannya. Salah satu contohnya adalah pada tahap analisis kueri (query analysis). Analisis kueri mengidentifikasi intensi pengguna saat mengetikkan kata kunci pada mesin pencari kemudian memberikan informasi yang relevan. Misalnya, jika kita mengetik nama seorang tokoh terkenal pada mesin pencari, akan muncul kotak bertuliskan biografi mengenai tokoh tersebut, lengkap dengan berbagai informasi penting lainnya. Selain itu juga muncul beberapa pertanyaan yang sering ditanyakan mengenai tokoh tersebut.

Selain itu, mesin pencari juga memiliki fungsi koreksi dan rekomendasi kueri. Fungsi ini biasanya muncul saat kita mengetikkan ejaan yang salah dalam mesin pencari. Ketika hal tersebut terjadi, mesin menunjukkan koreksi dengan label seperti menampilkan hasil untuk (ejaan yang dikoreksi) atau telusuri (ejaan salah yang kita ketik). Fungsi ini tentu memudahkan proses pencarian dan mengoptimalkan pengalaman pengguna.

# Teknik Natural Language Processing

Seperti yang telah dijelaskan pada materi sebelumnya, sebuah aplikasi NLP menerapkan berbagai teknik untuk menyelesaikan permasalahan yang berbeda. Pada bagian ini, kita akan membahas secara singkat beberapa tugas inti atau teknik NLP yang sering digunakan dalam aplikasi NLP. 

*   Pembuatan Teks (Text generation) <br>
Pembuatan teks, disebut juga *Natural Language Generation* (NLG) adalah proses menghasilkan keluaran teks bahasa alami dari bahasa lain. Teknik ini juga sering disebut sebagai *text-to-text generation*. Sebagai contoh, mesin penerjemah, yang telah kita bahas sebelumnya merupakan salah satu penerapan pembuatan teks. Selain itu, penyederhanaan teks, pembuatan ringkasan, serta koreksi kesalahan tata bahasa juga merupakan contoh penerapan dari teknik ini. Semua tugas ini menghasilkan teks bahasa alami sebagai keluaran.

*   Pencarian Informasi (Information Retrieval) <br>
Pencarian informasi memiliki arti yang sangat luas. Membuka kontak dalam ponsel untuk menemukan nomor telepon seseorang adalah bentuk pencarian informasi. Namun, dalam bidang akademis, pencarian informasi dapat didefinisikan sebagai berikut:

Information retrieval (IR) adalah proses menemukan informasi (biasanya berupa dokumen atau teks) yang relevan dengan kebutuhan. Informasi ini diambil dari kumpulan sumber informasi yang tersedia (biasanya di komputer atau internet) [33](https://nlp.stanford.edu/IR-book/html/htmledition/boolean-retrieval-1.html).
Aplikasi seperti mesin pencari yang telah kita bahas pada materi sebelumnya adalah contoh penggunaan teknik pencarian informasi.

*   Pemodelan Bahasa <br>
Dalam pemodelan bahasa, kita memprediksi kata berikutnya dalam kalimat berdasarkan histori kata-kata sebelumnya. Tujuannya untuk mempelajari kemungkinan urutan kata yang muncul. Pemodelan bahasa digunakan untuk menemukan solusi berbagai masalah, antara lain pengenalan suara, pengenalan tulisan tangan, terjemahan mesin, dan koreksi ejaan atau tata bahasa.

*   **Klasifikasi Teks** <br>
Ini merupakan proses membagi teks ke dalam kumpulan kategori tertentu berdasarkan isinya. Klasifikasi teks merupakan teknik paling populer di NLP dan digunakan dalam berbagai aplikasi, misalnya identifikasi spam email dan analisis sentimen. Inilah teknik yang akan kita pelajari pada modul ini.

# Tantangan Natural Language Processing

Pada materi sebelumnya, kita telah membahas mengenai berbagai contoh aplikasi NLP dan teknik-teknik yang digunakan dibalik aplikasi tersebut. Namun, tahukah Anda bahwa NLP dikenal sebagai bidang yang cukup menantang? Apa penyebabnya? Ambiguitas, kreativitas, dan keragaman bahasa alami merupakan tiga contoh karakteristik yang menjadi tantangan dalam pengembangan NLP [31](http://www.practicalnlp.ai/).

* Ambiguitas <br>
Ambiguitas menurut KBBI adalah kemungkinan adanya lebih dari satu makna dalam kata, gabungan kata, atau kalimat. Ia sering disebut juga sebagai ketidakpastian makna. Untuk lebih jelasnya, mari perhatikan beberapa contoh kalimat yang disadur dan diterjemahkan dari Winograd Schema Challenge, tulisan Prof. Terry Winograd, Universitas Stanford [34](https://www.aaai.org/ocs/index.php/KR/KR12/paper/viewFile/4492/4924).<br><br>
Dia tidak bisa mengangkat anaknya karena sangat **lemah**.
Pertanyaannya, siapa yang lemah?<br><br>
Dia tidak bisa mengangkat anaknya karena sangat **berat**.
Pertanyaannya, siapa yang berat?<br><br>
John **berjanji** pada Bill untuk pergi, jadi sejam kemudian dia pergi.<br>
John **menyuruh** Bill untuk pergi, jadi sejam kemudian dia pergi.<br>
Pertanyaan untuk kedua kalimat ini, siapa yang pergi?<br><br>
Skema ini memiliki pasangan kalimat yang hanya berbeda beberapa kata (ditebalkan). Namun, perbedaan ini menyebabkan makna kalimatnya sering terbalik. Sebagai manusia, tentu mudah untuk memahami kalimat-kalimat tersebut. Namun sayangnya, tidak bagi mesin. Ambiguitas seperti ini berpengaruh terhadap keluaran sistem. Jelas ini merupakan tantangan tersendiri dalam bidang NLP.

* Keragaman Bahasa<br>
Tahukah Anda? Berdasarkan data dari Badan Bahasa Indonesia [35](https://www.kemdikbud.go.id/main/blog/2018/07/badan-bahasa-petakan-652-bahasa-daerah-di-indonesia), terdapat 652 bahasa di Indonesia. Keragaman ini membuat pemetaan kosakata antar bahasa menjadi cukup rumit untuk dilakukan. Dampaknya, transfer solusi NLP dari satu bahasa ke bahasa lain juga menjadi sulit. Solusi untuk satu bahasa mungkin tidak berfungsi sama sekali untuk bahasa lainnya. Sehingga, kita perlu membangun solusi terpisah untuk setiap bahasa. Tentu ini melelahkan dan memakan waktu.

* Kreativitas<br>
Selain keragaman antar bahasa, aspek kreativitas dalam satu bahasa juga merupakan tantangan bagi pengembangan bidang NLP. Beberapa contoh aspek kreativitas dalam bahasa alami antara lain, variasi, gaya bahasa, dialek, genre, dan serapan bahasa daerah. Tentunya tidak mudah untuk membuat mesin bisa memahami berbagai aspek kreativitas ini. <br>

# Natural Language Processing Pipeline

Seperti sub-bidang machine learning lainnya, proyek NLP juga memiliki banyak tahapan mulai dari pengumpulan data hingga penerapan dan pemantauan model. Sebagian tahapan berlaku untuk semua alur (pipeline) ML, sedangkan sebagian lainnya bersifat khusus dan spesifik untuk NLP. 

Ingatlah bahwa setiap kasus dalam proyek machine learning bersifat unik. Namun, tentu ada alur atau prosedur yang umum digunakan dalam NLP pipeline. Nah, memahami prosedur umum dalam alur proyek NLP akan memudahkan kita dalam mengerjakan proyek NLP di kemudian hari. 

Secara umum, komponen utama alur proyek pengembangan sistem NLP adalah sebagai berikut [31] (http://www.practicalnlp.ai/).
image.png

Langkah pertama dalam proses pengembangan sistem NLP adalah proses pengumpulan data. Biasanya, data yang berhasil dikumpulkan tidak selalu bersih. Di sinilah proses pembersihan teks berperan. Setelah dibersihkan, data teks seringkali masih memiliki banyak variasi dan perlu diubah menjadi bentuk standar. Pada tahap inilah kita melakukan pra-pemrosesan. 

Selanjutnya masuk ke tahapan rekayasa fitur, tujuannya adalah menemukan indikator yang cocok untuk tugas kita. Indikator-indikator ini diubah menjadi format yang dapat dimengerti oleh algoritma pemodelan. 

Seperti proyek machine learning pada umumnya, tahapan selanjutnya adalah fase pemodelan, evaluasi, penerapan model dalam produksi (deployment), monitoring dan pembaruan model. Tahapan ini kurang lebih sama seperti alur proyek lain.

Namun, prosesnya di dunia nyata mungkin tidak selalu linier seperti yang ditunjukkan pada gambar di atas. Sering kali, NLP pipeline melibatkan beberapa tahapan bolak-balik dan perulangan (loop). Perulangan yang paling sering adalah dari tahap evaluasi ke tahap pra-pemrosesan, rekayasa fitur, pemodelan, dan kembali ke evaluasi. Ada juga perulangan secara keseluruhan mulai dari proses monitoring, akuisisi data, dst. Perulangan ini biasanya terjadi di tingkat proyek.

Pada fase awal, sebagian besar waktu digunakan dalam pemodelan dan evaluasi, sedangkan setelah sistem matang, rekayasa fitur membutuhkan waktu lebih lama. Pada modul ini, kita akan membahas beberapa tahapan yang spesifik untuk NLP saja, antara lain:

* Data Akuisisi
* Teks Cleaning dan Pre-Processing
* Feature Engineering.

# Akuisisi Data 

Data! Data! data!" he cried impatiently. "I can't make bricks without clay.

~Arthur Conan Doyle pada novel Sherlock Holmes: The Adventure of the Copper Beeches.

Sherlock Holmes adalah detektif yang hebat. Ia terkenal karena mampu melihat data yang mungkin terlewatkan oleh orang lain. Lumpur di sepasang sepatu, kemiripan dalam keluarga, bahkan sehelai rambut atau bulu yang menempel di jas atau gaun kliennya. Konsep ini --mengumpulkan, melihat, dan mengamati data sebelum membuat kesimpulan akan suatu hal-- sejatinya telah ada dalam berbagai cerita pendek sejak ratusan tahun lalu. 

Relevan dengan itu, data adalah inti dari setiap sistem machine learning. Keberhasilan suatu sistem machine learning sangat dipengaruhi oleh kualitas dan jumlah data yang dimiliki oleh sistem. Sehingga, tugas kita adalah mengumpulkan, mengamati, menganalisis, dan memproses data terlebih dahulu sebelum memutuskan untuk memasukkannya ke dalam sistem. 

Dalam NLP, kadang kita memerlukan jenis data yang berbeda. Misalnya, data teks dengan label, data real-time, maupun data historis selama beberapa tahun ke belakang. Pertanyaannya, bagaimana kita mendapatkan data itu

Terdapat beberapa cara akuisisi data, antara lain:


## Menggunakan Data Publik

Bagi Anda yang telah lama berkecimpung di bidang data, tentu tidak asing lagi dengan situs penyedia data seperti [Kaggle](https://www.kaggle.com/datasets), [UCI](https://archive.ics.uci.edu/ml/index.php), [mesin pencari dataset Google](https://datasetsearch.research.google.com/), maupun GitHub. Tahukan Anda? Selain sebagai tools version control dan ruang kolaborasi, para developer menggunakan GitHub untuk berbagi informasi mengenai berbagai sumber dataset untuk NLP. Contohnya adalah repository dari [Nicolas Iderhoff](https://github.com/niderhoff/nlp-datasets) dan [IndoNLU](https://github.com/indobenchmark/indonlu/tree/master/dataset).

Selain itu, setiap layanan cloud seperti Google Cloud Service, Azure, maupun AWS juga menyediakan dataset publik yang bisa diakses oleh para penggunanya. Jika menemukan dataset publik yang relevan dan sesuai dengan pekerjaan yang sedang kita lakukan, gunakan dataset tersebut, buat modelnya, dan lakukan evaluasi.

Permasalahannya, jika tidak menemukan data yang sesuai, apa yang bisa kita lakukan untuk mendapatkan data?


## Mengumpulkan Data dari Internet (Web Scraping)

Internet merupakan sumber data yang melimpah. Web scraping adalah metode untuk memproses dan mengumpulkan data dari berbagai situs di internet. Dengan metode tersebut, Anda dapat mengekstrak konten dan informasi untuk analisis lebih lanjut. 

Para praktisi data biasanya melakukan proses scraping pada situs media sosial seperti twitter, wikipedia, bahkan e-commerce. Analisis data dari proses web scraping dapat digunakan untuk berbagai tujuan, antara lain: keperluan riset, memantau sentimen pasar, dan mendapatkan informasi mengenai kompetitor. Beberapa tools yang dapat digunakan untuk proses ini antara lain [Beautifulsoup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/), [Selenium](https://www.selenium.dev/), dan [Scrapy](https://scrapy.org/). 


## Data Augmentation

Pada bidang computer vision, image augmentation adalah teknik standar yang sering digunakan. Sementara itu, pada NLP, text augmentation (augmentasi teks) masih jarang digunakan. Operasi pada gambar seperti proses memperbesar, membalik (flip) dan memutar (rotate) tidak mengubah gambar aslinya. Sehingga, proses transformasi melalui image augmentation merupakan perangkat penting dalam bidang computer vision.

Berbeda dengan computer vision, transformasi dengan text augmentation pada NLP menghadapi beberapa tantangan. Dua diantaranya adalah tantangan perubahan semantik (makna) dan tantangan kompleksitas bahasa yang tinggi seperti adanya konteks pembicaraan.

Meskipun demikian, terdapat beberapa pendekatan atau metode text augmentation yang bisa dilakukan untuk menghasilkan lebih banyak data. Disadur dari Chaudhary dalam tulisannya yang berjudul Data Augmentation for NLP [36](https://amitness.com/2020/05/data-augmentation-for-nlp/), berikut adalah beberapa metode text augmentation:

* Substitusi Leksikal (Lexical Substitution)
Substitusi leksikal merupakan teknik yang bekerja dengan mengganti kata pada teks tanpa mengubah arti dalam kalimat. Teknik ini sendiri memiliki beberapa pendekatan, dua diantaranya adalah Thesaurus-based substitution (substitusi berdasarkan tesaurusnya) dan Word-Embeddings Substitution (substitusi dengan teknik word-embeddings). Jangan bingung dulu dengan istilah-istilah ini ya, kita akan bahas satu per satu.
 1.   Thesaurus-based substitution \
Dengan teknik ini, kita mengambil kata acak dari kalimat kemudian mengganti dengan sinonimnya menggunakan [tesaurus](https://kbbi.kemdikbud.go.id/entri/tesaurus). Sebagai contoh, untuk bahasa inggris, kita bisa menggunakan database [WordNet](https://wordnet.princeton.edu/) dalam mencari sinonimnya, kemudian melakukan penggantian. WordNet merupakan database yang dikurasi secara manual dengan hubungan antar kata.

 2.   Word-Embeddings Substitution \
Dalam teknik ini, kita mengambil pre-trained word embeddings seperti Word2Vec, GloVe, FastText, Sent2Vec. Kemudian, kita menggunakan kata terdekat pada word embedding tersebut sebagai pengganti beberapa kata dalam kalimat. Sebagai contoh, kita bisa mengganti kata berikut dengan 3 kata yang paling mirip artinya untuk mendapatkan variasi pada teks.

* Back Translation
Dalam teknik ini, kita menggunakan aplikasi mesin penerjemah untuk memparafrasakan teks sambil melatih kembali maknanya. [Xie dkk](https://arxiv.org/abs/1904.12848) menggunakan metode ini untuk menambah teks yang tidak berlabel dan mempelajari model semi-supervised pada dataset IMDB dengan hanya 20 contoh berlabel. Model mereka mengungguli model sebelumnya yang dilatih pada 25.000 contoh berlabel. \
Bayangkan, hanya dengan 20 contoh model berlabel mampu mengungguli 25.000 contoh berlabel. Luar biasa!

Sampai di sini tentu Anda penasaran, seperti apa sih proses back-translation ini? Nah, tahapan proses back-translation adalah sebagai berikut:

* Ambil beberapa kalimat (misalnya dalam bahasa Inggris) dan terjemahkan ke bahasa lain, misalnya Perancis

* Terjemahkan kalimat bahasa Prancis menjadi kalimat bahasa Inggris kembali

* Periksa apakah kalimat baru berbeda dari kalimat asli. Jika ya, maka kita akan menggunakan kalimat baru ini sebagai versi tambahan dari teks aslinya.

Anda juga dapat menjalankan terjemahan balik menggunakan beberapa bahasa yang berbeda sekaligus untuk menghasilkan lebih banyak variasi. Hal ini membuat prosesnya menjadi lebih efektif dan tentu saja dapat menekan sumber daya. Mari kita perhatikan contoh gambar di bawah.

# Text Cleaning dan Pre-processing

Pembersihan teks dan pre-processing merupakan tahapan data wrangling dalam proyek NLP. Ini adalah proses mengekstrak teks mentah (raw text) dari data masukan, menghapus semua informasi yang tidak diperlukan, dan mengonversi teks ke dalam format yang dibutuhkan. Sehingga, teks siap dimasukkan ke dalam sistem NLP. 

Secara umum, terdapat tiga komponen utama pada tahap ini, yaitu:

* Menghilangkan Noise (gangguan) pada Teks
* Proses Tokenisasi
* Proses Normalisasi

### Menghilangkan Noise

Bayangkan Anda melakukan proses scraping dari situs wikipedia. Setelah mendapatkan datanya, Anda menyadari data tersebut dalam format html. Isi data mengandung banyak noise yang tidak relevan dengan fitur yang ingin Anda ekstrak dari teks. Anda dapat menghapus semua noise (informasi yang tidak perlu) tersebut agar menjadi teks yang utuh maknanya.

Beberapa tahapan untuk menghapus noise antara lain:

* Remove HTML Tag \
Teks tidak terstruktur biasanya mengandung banyak noise. Apalagi jika Anda menggunakan teknik seperti web scraping atau screen scraping untuk mengambil data dari halaman web. Tag HTML dan JavaScript biasanya tidak memiliki makna untuk proses analisis teks. Tujuan kita adalah mengekstrak konten tekstual yang bermakna dari data yang diambil dari web. Oleh karena itu, kita perlu melakukan proses penghapusan tag html.

* Spelling Correction \
Siapa di sini yang masih melakukan kesalahan penulisan alias typo? Jangan khawatir, sebagian besar dari kita mungkin masih melakukan kesalahan semacam itu. Pengetikan cepat pada keyboard komputer maupun layar ponsel cerdas berpotensi memiliki kesalahan ejaan penulisan. Hal ini sangat lazim terjadi. Jika dibiarkan, ini bisa menjadi permasalahan yang dapat merusak pemahaman linguistik data.

* Remove Stop Words \
Stop words adalah kata-kata yang memiliki sedikit makna pada teks. Biasanya, stop words ini berupa kata-kata yang paling sering muncul pada sekumpulan teks (corpus). Ia bisa berbentuk artikel, konjungsi, preposisi, dan sebagainya. Beberapa contoh stopword dalam bahasa inggris adalah a, an, is, and, dan the. \
Stop words harus dihapus sehingga algoritma machine learning dapat lebih fokus pada kata-kata yang memberikan makna dari teks. Kita bisa menggunakan daftar yang terdapat pada [nltk.corpus](https://www.nltk.org/) packages untuk melakukan proses penghapusannya. Selain itu, kita juga bisa menambah atau menghapus kata-kata khusus menyesuaikan dengan proyek dan teks yang sedang kita kerjakan.

* Remove Special Character \
Special character (karakter khusus) adalah karakter selain alfanumerik. Karakter ini paling sering ditemukan dalam referensi, komentar, nilai tukar mata uang, dsb. Karakter ini tidak menambah maka pada pemahaman teks sehingga dapat menyebabkan noise ke dalam algoritma. Untuk mengatasi dan menghilangkannya, kita bisa menggunakan regular expression atau regex.


### Tokenisasi

Tokenisasi adalah tahap membagi teks yang lebih panjang menjadi potongan-potongan yang lebih kecil (biasa disebut token). Token adalah komponen independen yang memiliki makna tertentu. Dengan begini, setiap kata yang memiliki makna berbeda akan diperlakukan berbeda oleh model. 

Tokenisasi dibagi menjadi dua cara, yaitu tokenisasi kalimat dan tokenisasi kata. Tokenisasi kalimat adalah proses memecah dokumen (corpus) menjadi kalimat. Ia juga dikenal sebagai segmentasi teks karena di sini kita membagi teks menjadi kalimat yang bermakna. Ada beberapa cara untuk melakukan tokenisasi kalimat. Cara yang paling sederhana adalah dengan mencari pembatas antar kalimat seperti titik (.) atau tanda baca lain seperti tanda tanya (?). 

Sementara itu, tokenisasi kata adalah proses pemisahan kalimat menjadi kata-kata penyusunnya. Tokenisasi kata memainkan peran penting dalam beberapa proses NLP, seperti saat proses pembersihan teks dan proses normalisasi. Terdapat beberapa teknik untuk melakukan tokenisasi kata, antara lain:

* Word_tokenize 
* TreebankWordTokenizer
* RegexpTokenizer

### Normalisasi Teks
Anda mungkin pernah belajar mengenai teknik normalisasi pada data numerik. Bagaimana kita melakukan normalisasi pada data teks? Inilah yang akan kita bahas pada materi kali ini. 

Sebuah kata dapat memiliki beberapa imbuhan. Sebuah angka kadang dituliskan dalam bentuk angka, kadang juga dalam bentuk teks. Misal, angka 9 kadang ditulis sebagai sembilan atau 9. Dalam bahasa inggris, kadang kata dieja dengan cara yang berbeda atau memiliki bentuk yang berbeda. Sebagai contoh, is, am, are memiliki bentuk dasar be. Nah, normalisasi teks adalah mengonversi teks menjadi suatu bentuk yang mencakup semua variasi.

Beberapa contoh normalisasi pada teks antara lain:

* Mengubah semua teks menjadi huruf kecil atau huruf besar, 
* Mengonversi semua angka menjadi teks, 
* Mengeliminasi imbuhan, 
* Mengubah bentuk kata menjadi bentuk dasarnya, dsb. 

Ada beberapa cara untuk melakukan normalisasi. Dua diantaranya yaitu proses stemming, dan lemmatization. Stemming adalah proses mengeliminasi imbuhan untuk mendapatkan kata dasarnya. \
Sedangkan lemmatization adalah proses mengubah kata menjadi bentuk dasar yang bermakna dengan melibatkan konteks. Ia mengelompokkan kata-kata yang memiliki bentuk dasar yang sama sehingga dapat diidentifikasikan menjadi satu item. Bentuk dasar disebut sebagai lema kata dan kadang juga dikenal sebagai bentuk baku.

Selain stemming dan lemmatization, proses normalisasi lainnya adalah:

* Mengubah semua karakter menjadi huruf kecil.
* Mengubah penulisan angka dalam bentuk teks.

# Feature Engineering untuk Representasi Teks

Data teks memerlukan persiapan khusus sebelum digunakan untuk pemodelan. Teks harus diurai terlebih dahulu dengan teknik tokenisasi yang telah Anda pelajari pada materi sebelumnya. Kemudian, ia perlu dikodekan menjadi bilangan numerik agar dapat digunakan sebagai masukan pada algoritma machine learning. Proses ini disebut ekstraksi atau rekayasa fitur (disebut juga sebagai vektorisasi).

Pada materi kali ini, kita akan belajar mengenai teknik rekayasa fitur untuk representasi teks. Teknik ini mengubah teks pada data masukan ke dalam bentuk numerik sehingga dapat digunakan untuk pemodelan. Ia merupakan salah satu tahapan penting dalam menentukan hasil model NLP.

Apa pun jenis datanya, baik data teks, gambar, suara, maupun video, representasi fitur adalah tahapan umum dalam setiap proyek machine learning. Namun, representasi fitur untuk data teks seringkali jauh lebih menantang dibandingkan dengan format data lainnya. Untuk memahami hal ini, mari kita simak bagaimana format data gambar dan video dapat direpresentasikan secara numerik. 

Ilustrasi di atas menunjukkan bagaimana mata kita melihat seekor kucing dan bagaimana mesin merepresentasikan fitur pada gambar kucing. Komputer membaca gambar dalam bentuk matriks piksel dengan nilai piksel mewakili intensitas warna. Representasi matriks ini secara akurat mewakili seluruh gambar dengan lengkap.

Demikian juga dengan video. Pada dasarnya, video adalah kumpulan frame dalam jangka waktu tertentu, di mana setiap frame-nya adalah gambar. Oleh karena itu, setiap video juga direpresentasikan sebagai kumpulan matriks yang berurutan untuk setiap frame-nya.

Sampai sini jelas ya, kita dapat langsung merepresentasikan gambar dan video secara matematis. Pertanyaan selanjutnya adalah, bagaimana merepresentasikan data teks secara matematis?

Terdapat beberapa metode rekayasa fitur untuk representasi teks. Metode ini diklasifikasikan ke dalam empat kategori [31](http://www.practicalnlp.ai/), antara lain:
* Basic vectorization approaches (contoh: One-Hot Encoding, Bag of Words, Bag of N-Grams, dan TF-IDF).
* Distributed representations (contoh: Words Embeddings).
* Universal text representation (menggunakan pre-trained embedding).
* Handcrafted features (mengandalkan domain-specific knowledge).

Pada modul ini, kita akan membahas dua metode populer dari kategori basic vectorization approaches, yaitu:

* Bag of Words
* TF-IDF

Dari kedua metode tersebut, kita akan menggunakan TF-IDF dan menerapkannya dalam proyek Sentimen Analisis. Mari kita bahas kedua metode ini satu per satu.

### Bag of Words
Bag of Words (BoW) adalah teknik representasi teks yang umum digunakan terutama dalam klasifikasi teks. Ia merepresentasikan teks sebagai kantong (kumpulan) kata-kata dengan mengabaikan urutan dan konteksnya. Ia dinamai demikian (kantong kata-kata), karena model hanya memperhatikan kemunculan kata-kata yang ada dalam dokumen. Ia tidak mempertimbangkan di mana kata-kata tersebut muncul dan bagaimana urutan atau struktur kata dalam dokumen. 

Intuisi dasar di balik BoW adalah asumsi bahwa kelas tertentu dalam data memiliki ciri kumpulan kata yang unik. Jika dua potongan teks memiliki kata-kata yang hampir sama, mereka termasuk dalam tas (kelas) yang sama. Jadi, dengan menganalisis kata-kata yang ada dalam sepotong teks, seseorang dapat mengidentifikasi kelas (tas) yang dimilikinya [31](http://www.practicalnlp.ai/).

Cukup jelas, ya? 

Untuk memahami BoW dengan lebih jelas, perhatikan contoh teks yang akan kita gunakan sebagai data berikut ini.

Ini adalah hari yang aku tunggu-tunggu. Hari ini aku diwisuda. Rasanya menyenangkan bisa diwisuda bersama teman-teman.

Jika kalimat di atas dihitung, kita memiliki 18 data kata. Tahap selanjutnya adalah membuat daftar kosa kata dalam data. Kata unik dalam data di atas (mengabaikan huruf besar dan tanda baca) adalah sebagai berikut:

ini 
adalah 
hari 
yang
aku
tunggu 
diwisuda 
rasanya 
menyenangkan 
bisa 
bersama 
teman 

Daftar kosa kata ini sekarang memiliki 12 kata. Selain memperhatikan kosa kata yang telah dikenal sebelumnya, BoW juga mempertimbankan ukuran kemunculan kosa kata yang telah dikenali tersebut.

Hal ini dilakukan pada tahap selanjutnya yaitu membuat skor setiap kata pada dokumen untuk menghasilkan vektor. Cara menghitung skor yang paling mudah adalah dengan bilangan biner. Vektornya bernilai 1 jika kata tersebut muncul dan bernilai 0 jika tidak. Sehingga, vektornya menjadi sebagai berikut.

ini = 1
adalah = 0 
hari = 1
yang = 0
aku = 1
tunggu = 1
diwisuda = 1
rasanya = 0
menyenangkan = 0
bisa = 0
bersama = 0
teman = 1
Jika kita representasikan bilangan biner tersebut ke dalam vektor, maka hasilnya sebagai berikut.

[1, 0, 1, 0, 1, 1, 1, 0, 0, 0, 0, 1]

Sehingga, masing-masing kalimat pada data kita akan berbentuk vektor seperti ini:

Ini adalah hari yang aku tunggu-tunggu = [1, 0, 1, 0, 1, 1]

Hari ini aku diwisuda = [1, 1, 1, 1]

Rasanya menyenangkan bisa diwisuda bersama teman-teman = [0, 0, 0, 1, 0, 1]

Dari contoh sederhana ini kita dapat melihat bahwa hanya kemunculan kata-kata yang diketahui saja yang diberi skor, sementara kata-kata yang tidak dikenal (hanya muncul sekali) diabaikan. Seiring dengan semakin besarnya ukuran dokumen, ukuran vektor juga akan semakin panjang. 

Cukup simpel dan mudah dimengerti, kan? 

Namun, teknik pemberian skor ini memiliki kekurangan. Kata-kata yang sangat sering muncul dalam dokumen mungkin tidak mengandung konten informatif terhadap model. Sebagai contoh, kata the dalam teks bahasa inggris. Sedangkan, kata-kata yang jarang muncul justru mungkin mengandung makna penting dalam teks.
